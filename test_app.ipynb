```python
import cv2
import base64
import requests
import numpy as np
from matplotlib import pyplot as plt

url = 'http://0.0.0.0:2103/solve?'
img_bytes = requests.get('https://alma-ata.kdmid.ru/queue/CodeImage.aspx?id=c270').content

def readb64(uri):
    nparr = np.fromstring(uri, np.uint8)
    img = cv2.imdecode(nparr, cv2.IMREAD_COLOR)
    return img

image = readb64(img_bytes)
retval, buffer = cv2.imencode('.jpg', image)
```

```python
plt.imshow(image)
```

```python
params = {'image': base64.b64encode(img_bytes)}
resp = requests.get(url, params=params)
```
